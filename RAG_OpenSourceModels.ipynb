```python
from langchain_together.embeddings import TogetherEmbeddings
from langchain.vectorstores import Chroma
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.chains import RetrievalQA
from langchain_community.document_loaders import PyPDFLoader
from langchain.embeddings import HuggingFaceBgeEmbeddings


```

The embedding models on Together AI are not as good as the ones on Hugging face, which outperform GPTs embedding models. I have used Hugging Face's BGE embeddings <br>
https://python.langchain.com/docs/integrations/text_embedding/bge_huggingface <br>

```python
model_name = "BAAI/bge-base-en"
encode_kwargs = {'normalize_embeddings': True} # set True to compute cosine similarity

model_norm = HuggingFaceBgeEmbeddings(
    model_name=model_name,
    model_kwargs={'device': 'cuda'},
    encode_kwargs=encode_kwargs
)
```

```python
loader = PyPDFLoader("/content/CS_982_Assignment.pdf")
pages = loader.load()
```

```python
text_splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=200)
texts = text_splitter.split_documents(pages)
```

```python
len(texts)
```

```python
persist_directory = 'vectordb'

## Here is the nmew embeddings being used
embeddings = model_norm

vectordb = Chroma.from_documents(documents=texts,
                                 embedding=embeddings,
                                 persist_directory=persist_directory)
```

```python
query = "Are airlines industry profitable"
docs = vectordb.similarity_search(query)
for doc in docs:
    print (f"{doc.page_content}\n")
```

```python
retriever = vectordb.as_retriever(search_kwargs={"k": 3})
```

```python
docs2 = retriever.get_relevant_documents("what are the top services in the flight")
for doc in docs2:
    print (f"{doc.page_content}\n")
```

Using the following links, the right text for the model was obtained for Llama 70B model
https://python.langchain.com/docs/integrations/llms/together <br>  https://docs.together.ai/docs/inference-models

```python
from langchain_together import Together
llm = Together(
    model="togethercomputer/llama-2-70b-chat",
    temperature=0.3,
    max_tokens=128,
    top_k=1,

)
```

```python
qa_chain = RetrievalQA.from_chain_type(llm=llm,
                                  chain_type="stuff",
                                  retriever=retriever,
                                  return_source_documents=True)
```

```python
query = "what are the top services in the flight"
llm_response = qa_chain(query)
llm_response["result"]
```

```python
query = "are airlines industries profitable"
llm_response = qa_chain(query)
llm_response["result"]
```
